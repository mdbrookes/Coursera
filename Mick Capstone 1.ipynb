```python
#
#      FIRST PART OF ASSIGNMENT
#
#

import numpy as np
# library to handle data in a vectorized manner

import pandas as pd 
# library for data analsysis

import requests
from bs4 import BeautifulSoup
# import beautifule soud

#Create varaibke points in to ("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

# Extract tables
dfs = pd.read_html(url)
df=  dfs[0]

# Create Dataframe with labels
df2 = df[['Postal Code', 'Borough','Neighborhood']]


# Sort buy  postal code
df2.sort_values(by='Postal Code', inplace=True)

#Set up blank variable
Postal_code='xxx'
Neighbourhood='xxx'

# Go through whole dataframe
for i in range (len(df2)):

# Rule is : More than one neighborhood can exist in one postal code area. For example, in the table on the Wikipedia page, you will notice that
#M5A is listed twice and has two neighborhoods: Harbourfront and Regent Park. These two rows will be combined into one row with the 
#neighborhoods separated with a comma as shown in row 11 in the above table.
# Copy last row's neighnourhood into this row's neighbourhood if they sahre postal code 
    if df2.iloc[i,0]==Postal_code:
        df2.iloc[i,2]= df2.iloc[i,2] + ', ' + Neighbourhood
        df2.drop([i])
    Postal_code=df2.iloc[i,0]
    Neighbourhood=df2.iloc[i,2]

# Rule is : If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.
# Replace Neihbourhood(col1) if NOt assigned with Borough name (col2)
    if df2.iloc[i,2] =='Not assigned':
        df2.iloc[i,2]=df2.iloc[i,1]  
df2.drop_duplicates(subset="Postal Code", keep='last', inplace=True) 
    

#Rule is : "Only process the cells that have an assigned borough. Ignore cells with a borough that is "Not assigned."
# Create new dataframe df 3 with Borough NOt assigned
df3= df2[df2.Borough != 'Not assigned'] 

#sort by original index
df3.sort_index(inplace=True)

# Reset index
df3.reset_index(inplace=True,drop=True)



#Rseize the page 
pd.set_option('display.expand_frame_repr', False)
print(df3.head(105))

print (df3.shape)


#
#     THIS COMPLETES FIRST PART OF ASSIGNMENT
#
#     The notebook contimes on next line 
#
#
```

```python

#
#     SECOND PART OF ASSIGNMENT
#
#
#


latlong = pd.read_csv (r'http://cocl.us/Geospatial_data')

df_inner = pd.merge(df3, latlong, on='Postal Code', how='inner')

df_inner

#
#     THIS COMPLETES SECOND PART OF ASSIGNMENT
#
#     The notebook contimes on next line 
#
#
```

```python

#
#     THIRD PART OF ASSIGNMENT
#
#
#
#!conda install -c conda-forge folium=0.5.0
import folium 

# import k-means from clustering stage
from sklearn.cluster import KMeans

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors


    
# set number of clusters to 4, use only longtitude and Latittude for clustering
kclusters = 4

df_inner_clustering = df_inner.drop(['Postal Code','Neighborhood', 'Borough'], axis=1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(df_inner_clustering)

# insert cluster labels to df_inner
#toronto_merge=df_inner.insert(0, 'Cluster Labels', kmeans.labels_)


#  create map of Toronto using latitude and longitude values
map_Toronto = folium.Map(location=[43.6487 , -79.38544], zoom_start=10)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df_inner['Latitude'], df_inner['Longitude'], df_inner['Neighborhood'], df_inner['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_Toronto)
       
# Sho Map       
map_Toronto

#
#     THIS COMPLETES THIRD PART OF ASSIGNMENT
#
#     The notebook contimes on next line 
#
#
```
